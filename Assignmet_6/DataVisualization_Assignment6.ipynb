### a) Use different visualizations to summarize the data (data exploration).

The pairwise relations between each variable are displayed in the off-diagonal graphs.
We can examine how the distribution of each variable varies across the various quality levels by using the hue parameter to colour the dots according to their level of quality. Although there is no obvious trend for pH, we can observe that higher quality wines often have greater alcohol contents and lower residual sugar levels.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

wine_data = pd.read_csv("/content/drive/My Drive/dataset/winequality-white.csv", sep=";")

sns.pairplot(wine_data, vars=['alcohol', 'pH', 'residual sugar'], hue='quality')
plt.show()
```

```python
wine_data = pd.read_csv("/content/drive/My Drive/dataset/winequality-red.csv", sep=";")
sugar_threshold = 2
above_threshold = wine_data[wine_data['residual sugar'] > sugar_threshold]['residual sugar'].count()
below_threshold = wine_data[wine_data['residual sugar'] <= sugar_threshold]['residual sugar'].count()
plt.pie([above_threshold, below_threshold], labels=['Above Threshold', 'Below Threshold'], autopct='%1.1f%%', startangle=90)
plt.title('Proportion of Red Wines with Residual Sugar  Above '+ str(sugar_threshold))

plt.show()
```

```python
wine_data = pd.read_csv("/content/drive/My Drive/dataset/winequality-white.csv", sep=";")
corr_matrix = wine_data.corr()
sns.heatmap(corr_matrix, annot=False, cmap="coolwarm")
plt.title('Correlation Matrix of Chemical Features for white wine')
plt.show()
```

```python
wine_data = pd.read_csv("/content/drive/My Drive/dataset/winequality-red.csv", sep=";")
corr_matrix = wine_data.corr()
sns.heatmap(corr_matrix, annot=False, cmap="coolwarm")
plt.title('Correlation Matrix of Chemical Features for red wine')
plt.show()
```

```python
wine_data = pd.read_csv("/content/drive/My Drive/dataset/winequality-white.csv", sep=";")
corr_matrix = wine_data.corr()
quality_corr = corr_matrix['quality']

plt.figure(figsize=(12, 8))
sns.barplot(x=quality_corr.index, y=quality_corr.values)
plt.title("Correlation between Wine Quality and Chemical Features for White Wine")
plt.xticks(rotation=45)
plt.show()
```

```python
wine_data = pd.read_csv("/content/drive/My Drive/dataset/winequality-red.csv", sep=";")
corr_matrix = wine_data.corr()
quality_corr = corr_matrix['quality']

plt.figure(figsize=(12, 8))
sns.barplot(x=quality_corr.index, y=quality_corr.values)
plt.title("Correlation between Wine Quality and Chemical Features for Red Wine")
plt.xticks(rotation=45)
plt.show()
```
